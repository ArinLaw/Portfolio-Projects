# FEATURE RELATIONSHIPS 

# Relationships between Multiple Variables(Pair plots, scatterplot, correlation matrix)

```python
#COMPARING TWO FEATURES SIDE BY SIDE USING A SCATTERPLOT
# Scatter plot of 'BMI' vs. 'Heart Rate'
plt.scatter(medical_data['BMI'], medical_data['Heart Rate'])
plt.xlabel('BMI')
plt.ylabel('Heart Rate')
plt.title('Scatter Plot: BMI vs. Heart Rate')
plt.show()
```

```python
#The scatter plot is too bulky due to a large number of data points. 
#We used random data sampling and data aggregating to reduce the number of observations plotted. 
#This will help in making the scatter plot more readable and easier to interpret.

# Group by 'BMI' and calculate the mean 'Heart Rate' for each 'BMI' group
aggregated_data = medical_data.groupby('BMI')['Heart Rate'].mean().reset_index()

# Scatter plot of 'BMI' vs. mean 'Heart Rate' using the aggregated data
plt.scatter(aggregated_data['BMI'], aggregated_data['Heart Rate'])
plt.xlabel('BMI')
plt.ylabel('Mean Heart Rate')
plt.title('Scatter Plot: BMI vs. Mean Heart Rate (Aggregated Data)')
plt.show()
```

```python
# Randomly sample 500 data points from the DataFrame
sampled_data = medical_data.sample(n=500, random_state=42)

# Scatter plot of 'BMI' vs. 'Heart Rate' using the sampled data
plt.scatter(sampled_data['BMI'], sampled_data['Heart Rate'])
plt.xlabel('BMI')
plt.ylabel('Heart Rate')
plt.title('Scatter Plot: BMI vs. Heart Rate (Sampled Data)')
plt.show()
```

```python
# Selecting some numerical features for correlation analysis
numerical_features = ['BMI', 'Heart Rate', 'Blood Pressure', 'Cholesterol']

# Create a correlation matrix
correlation_matrix = medical_data[numerical_features].corr()

# Plot a heatmap of the correlation matrix using Seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python

# Selecting the all numerical features for correlation analysis
numerical_features = ['Age','Weight','Height','BMI', 'Heart Rate', 'Blood Pressure', 'Cholesterol']

# Create a correlation matrix
correlation_matrix = medical_data[numerical_features].corr()

# Plot a heatmap of the correlation matrix using Seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#Pairplot shows pairwise relationships between multiple variables
# Randomly sample 1000 data points from the DataFrame
sampled_data = medical_data.sample(n=1000, random_state=42)

# Pairplot using the sampled data
sns.pairplot(sampled_data, diag_kind='kde')
plt.suptitle('Pairplot with Sampled Data', y=1.02)
plt.show()
```

```python
# Group by 'Blood Type' and calculate the mean values for numerical columns
aggregated_data = medical_data.groupby('Blood Type').mean().reset_index()

# Pairplot using the aggregated data
sns.pairplot(aggregated_data, diag_kind='kde')
plt.suptitle('Pairplot with Aggregated Data', y=1.02)
plt.show()
```

 # Comparison between Groups: - Compare numerical variables like BMI different groups like Gender 

# BMI & GENDER

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='BMI', data=medical_data)
plt.xlabel('Gender')
plt.ylabel('BMI')
plt.title('BMI Comparison between Male and Female')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='BMI', data=medical_data)
plt.xlabel('Gender')
plt.ylabel('BMI')
plt.title('BMI Comparison between Male and Female')

# Get the counts for each gender
counts = medical_data['Gender'].value_counts()

# Annotate each box with its count
for i, count in enumerate(counts):
    plt.text(i, medical_data['BMI'].max() + 1, f'Count: {count}', ha='center', va='center')

plt.show()
```

# BAR TYPE - BLOOD TYPE ANALYSIS

```python
# Count the occurrences of each blood type in the 'Blood Type' column
blood_type_counts = medical_data['Blood Type'].value_counts()

# Create a bar chart to visualize the distribution of blood types
plt.figure(figsize=(8, 6))
sns.barplot(x=blood_type_counts.index, y=blood_type_counts.values)
plt.xlabel('Blood Type')
plt.ylabel('Number of Students')
plt.title('Distribution of Students across Blood Types')
plt.show()
```

```python
# Count the occurrences of each blood type in the 'Blood Type' column
blood_type_counts = medical_data['Blood Type'].value_counts()

# Create a bar chart to visualize the distribution of blood types
plt.figure(figsize=(8, 6))
sns.barplot(x=blood_type_counts.index, y=blood_type_counts.values)
plt.xlabel('Blood Type')
plt.ylabel('Number of Students')
plt.title('Distribution of Students across Blood Types')

# Annotate each bar with its count
for i, count in enumerate(blood_type_counts.values):
    plt.text(i, count + 5, str(count), ha='center', va='center', fontweight='bold')

plt.show()
```

# Health Condition Analysis - Pie Chart

```python
# Count the occurrences of diabetes and smoking habits in the 'Diabetes' and 'Smoking' columns
diabetes_counts = medical_data['Diabetes'].value_counts()
smoking_counts = medical_data['Smoking'].value_counts()

# Create pie charts to visualize the percentage of students with diabetes and smoking habits
plt.figure(figsize=(12, 6))

# Pie chart for diabetes
plt.subplot(1, 2, 1)
plt.pie(diabetes_counts, labels=diabetes_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Percentage of Students with Diabetes')

# Pie chart for smoking habits
plt.subplot(1, 2, 2)
plt.pie(smoking_counts, labels=smoking_counts.index, autopct='%1.1f%%', colors=['lightgreen', 'lightsalmon'])
plt.title('Percentage of Students with Smoking Habits')

plt.tight_layout()
plt.show()
```
