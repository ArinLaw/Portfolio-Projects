```python
#Bar charts for 'education', 'self_employed', and 'loan_status'
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Bar chart for 'education'
ld['EDUCATION'].value_counts().plot(kind='bar', ax=axes[0])
axes[0].set_title('Education')
axes[0].set_ylabel('Count')

# Bar chart for 'self_employed'
ld['SELF_EMPLOYED'].value_counts().plot(kind='bar', ax=axes[1])
axes[1].set_title('Self Employed')
axes[1].set_ylabel('Count')

# Bar chart for 'loan_status'
ld['LOAN_STATUS'].value_counts().plot(kind='bar', ax=axes[2])
axes[2].set_title('Loan Status')
axes[2].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Bar charts for 'education', 'self_employed', and 'loan_status'
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Bar chart for 'education'
education_counts = ld['EDUCATION'].value_counts()
education_counts.plot(kind='bar', ax=axes[0])
axes[0].set_title('Education')
axes[0].set_ylabel('Count')

# Add count labels to 'education' bar chart
for i, count in enumerate(education_counts):
    axes[0].annotate(str(count), xy=(i, count), ha='center', va='bottom')

# Bar chart for 'self_employed'
self_employed_counts = ld['SELF_EMPLOYED'].value_counts()
self_employed_counts.plot(kind='bar', ax=axes[1])
axes[1].set_title('Self Employed')
axes[1].set_ylabel('Count')

# Add count labels to 'self_employed' bar chart
for i, count in enumerate(self_employed_counts):
    axes[1].annotate(str(count), xy=(i, count), ha='center', va='bottom')

# Bar chart for 'loan_status'
loan_status_counts = ld['LOAN_STATUS'].value_counts()
loan_status_counts.plot(kind='bar', ax=axes[2])
axes[2].set_title('Loan Status')
axes[2].set_ylabel('Count')

# Add count labels to 'loan_status' bar chart
for i, count in enumerate(loan_status_counts):
    axes[2].annotate(str(count), xy=(i, count), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Bar chart for 'loan_status'
loan_status_counts = ld['LOAN_STATUS'].value_counts()
ax = loan_status_counts.plot(kind='bar', figsize=(6, 4))
ax.set_title('Loan Status')
ax.set_ylabel('Count')

# Add count labels to 'loan_status' bar chart
for i, count in enumerate(loan_status_counts):
    ax.annotate(str(count), xy=(i, count), ha='center', va='bottom')

plt.show()
```

```python
# Pie chart for 'education'
education_counts = ld['EDUCATION'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(education_counts, labels=education_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Education Distribution')
plt.show()
```

```python
# Visualize numerical feature distributions using box plots
ld.boxplot(column=['INCOME_ANNUM', 'LOAN_AMOUNT', 'CIBIL_SCORE'], figsize=(10, 6))
plt.show()
```

```python
# Exploring relationships using correlation matrix
correlation_matrix = ld[['INCOME_ANNUM', 'LOAN_AMOUNT', 'RESIDENTIAL_ASSETS_VALUE','BANK_ASSET_VALUE','LUXURY_ASSETS_VALUE','COMMERCIAL_ASSETS_VALUE']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Select only the numeric columns for correlation calculation
numeric_columns = ld.select_dtypes(include=['int64', 'float64'])

# Calculate the correlation matrix
correlation_matrix = numeric_columns.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Exploring relationships using correlation matrix
correlation_matrix = ld[['INCOME_ANNUM', 'LOAN_AMOUNT', 'RESIDENTIAL_ASSETS_VALUE', 'BANK_ASSET_VALUE', 'LUXURY_ASSETS_VALUE', 'COMMERCIAL_ASSETS_VALUE']].corr()

# Create scatter plot for selected numerical columns
sns.pairplot(data=ld, vars=['INCOME_ANNUM', 'LOAN_AMOUNT', 'RESIDENTIAL_ASSETS_VALUE', 'BANK_ASSET_VALUE', 'LUXURY_ASSETS_VALUE', 'COMMERCIAL_ASSETS_VALUE'])
plt.suptitle('Scatter Plot for Selected Numerical Columns', y=1.02)
plt.show()
```

```python
# Select the numerical columns for scatter plots
numerical_columns = ['INCOME_ANNUM', 'LOAN_AMOUNT', 'RESIDENTIAL_ASSETS_VALUE', 'BANK_ASSET_VALUE', 'LUXURY_ASSETS_VALUE', 'COMMERCIAL_ASSETS_VALUE']

# Creating individual scatter plots for each combination of columns
plt.figure(figsize=(15, 12))
for i in range(len(numerical_columns)):
    for j in range(i+1, len(numerical_columns)):
        plt.subplot(3, 2, j-i)
        sns.scatterplot(data=ld, x=numerical_columns[i], y=numerical_columns[j])
        plt.xlabel(numerical_columns[i])
        plt.ylabel(numerical_columns[j])
plt.tight_layout()
plt.show()
```
